```python
%%time
from transformers import pipeline

#pipe = pipeline("text-to-speech", model = "")
pipe = pipeline("text-to-speech")
```

```python
text = "Hello, how are you today?"
output = pipe(text)
```

```python
# Audiosegment allows you to just pass in the numpy array in byte format and then export it.
from pydub import AudioSegment
```

```python
# If the voice is really slow and low, or high pitched and fast, your sampling rate is probably off!
# If you don't get this sampling rate correct, it may play it back too slowly or too fast.

# You can also do things like sample width and channels.
# So we can say the actual size of the sample is again the output.
# Audio array and then you can check the data type item size.
# You can also specify whether it's going to be stereo or mono.
# This code just exports a mono channel, so stereo would have different sounds on the left and right speaker.
audio_seg = AudioSegment(output['audio'].tobytes(), frame_rate=output['sampling_rate'],
                       sample_width = output['audio'].dtype.itemsize, channels = 1)
```

```python
audio_seg.export("output.mp3", format="mp3")# you can try wav
```
